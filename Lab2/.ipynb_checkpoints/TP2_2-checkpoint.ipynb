```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape and normalize the input data
x_train = x_train.reshape(-1, 28, 28, 1).astype('float32') / 255.0
x_test = x_test.reshape(-1, 28, 28, 1).astype('float32') / 255.0

# One-hot encode the labels
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)

# Split the training data into training and validation sets (80% training, 20% validation)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)

print("Training examples:", x_train.shape[0])
print("Validation examples:", x_val.shape[0])
print("Test examples:", x_test.shape[0])
```

```python
import tensorflow as tf

# Define the MLP architecture
model = tf.keras.Sequential([
    # Input layer
    tf.keras.layers.Flatten(input_shape=(28, 28, 1)),
    
    # First hidden layer with 128 neurons and ReLU activation
    tf.keras.layers.Dense(128, activation='relu'),
    
    # Second hidden layer with 64 neurons and ReLU activation
    tf.keras.layers.Dense(64, activation='relu'),
    
    # Fully connected layer with 10 neurons for the output and softmax activation
    tf.keras.layers.Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Print model summary
model.summary()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score

# Function to plot learning curves and return history
def plot_learning_curves(history, title, color):
    plt.figure(figsize=(12, 4))

    # Plot loss
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss', color=color)
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(title + ' - Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    # Plot accuracy
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy', color=color)
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title(title + ' - Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Return loss and accuracy for each epoch
    return history.history['loss'], history.history['val_loss'], history.history['accuracy'], history.history['val_accuracy']
```

```python
# Train the model with batchSGD optimizer
batchsgd_model = tf.keras.models.clone_model(model)
batchsgd_model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
batchsgd_history = batchsgd_model.fit(x_train, y_train, epochs=50, batch_size=len(x_train), validation_data=(x_val, y_val), verbose=0)
batchsgd_loss, batchsgd_val_loss, batchsgd_accuracy, batchsgd_val_accuracy = plot_learning_curves(batchsgd_history, 'batchSGD', color='red')
# Calculate F1 score, precision, and recall for each method
def calculate_metrics(y_true, y_pred):
    f1 = f1_score(y_true, y_pred, average='macro')
    precision = precision_score(y_true, y_pred, average='macro')
    recall = recall_score(y_true, y_pred, average='macro')
    return f1, precision, recall

batchsgd_predicted_labels = batchsgd_model.predict(x_test)
# Flatten the true and predicted labels
y_test_flat = np.argmax(y_test, axis=1)
batchsgd_predicted_labels_flat = np.argmax(batchsgd_predicted_labels, axis=1)
batchsgd_f1, batchsgd_precision, batchsgd_recall = calculate_metrics(y_test_flat, batchsgd_predicted_labels_flat)

print("batch SGD F1 Score:", batchsgd_f1)
print("batch SGD Precision:", batchsgd_precision)
print("batch SGD Recall:", batchsgd_recall)
```

```python
# Train the model with Mini-batch SGD optimizer
mini_batch_sgd_model = tf.keras.models.clone_model(model)
mini_batch_sgd_model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
                             loss='categorical_crossentropy',
                             metrics=['accuracy'])
mini_batch_sgd_history = mini_batch_sgd_model.fit(x_train, y_train, epochs=50, batch_size=64, validation_data=(x_val, y_val), verbose=0)

mini_batch_loss, mini_batch_val_loss, mini_batch_accuracy, mini_batch_val_accuracy = plot_learning_curves(mini_batch_sgd_history, 'Mini-batch SGD (Batch Size: 64)', color='black')
mini_batch_predicted_labels = mini_batch_sgd_model.predict(x_test)
mini_batch_predicted_labels_flat = np.argmax(mini_batch_predicted_labels, axis=1)
mini_batch_f1, mini_batch_precision, mini_batch_recall = calculate_metrics(y_test_flat, mini_batch_predicted_labels_flat)
print("Mini-batch SGD F1 Score:", mini_batch_f1)
print("Mini-batch SGD Precision:", mini_batch_precision)
print("Mini-batch SGD Recall:", mini_batch_recall)
```

```python
# Train the model with SGD optimizer
sgd_model = tf.keras.models.clone_model(model)
sgd_model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
                        loss='categorical_crossentropy',
                        metrics=['accuracy'])
sgd_history = sgd_model.fit(x_train, y_train, epochs=50, batch_size=1, validation_data=(x_val, y_val), verbose=0)

sgd_loss, sgd_val_loss, sgd_accuracy, sgd_val_accuracy = plot_learning_curves(sgd_history, 'SGD', color='green')
sgd_predicted_labels = sgd_model.predict(x_test)
sgd_predicted_labels_flat = np.argmax(sgd_predicted_labels, axis=1)
sgd_f1, sgd_precision, sgd_recall = calculate_metrics(y_test_flat, sgd_predicted_labels_flat)
print("SGD F1 Score:", sgd_f1)
print("SGD Precision:", sgd_precision)
print("SGD Recall:", sgd_recall)
```

```python
import time

# Function to train model and return training time
def train_model(model, optimizer, x_train, y_train, x_val, y_val, epochs, batch_size):
    start_time = time.time()
    history = model.fit(x_train, y_train, epochs=epochs, batch_size=batch_size, validation_data=(x_val, y_val), verbose=0)
    end_time = time.time()
    training_time = end_time - start_time
    return history, training_time

# Learning rate decay scheduler
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.01,
    decay_steps=1000,
    decay_rate=0.96,
    staircase=True)

# Train the model with Mini-batch SGD with decay
mini_batch_sgd_decay_model = tf.keras.models.clone_model(model)
mini_batch_sgd_decay_model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=lr_schedule),
                                   loss='categorical_crossentropy',
                                   metrics=['accuracy'])
mini_batch_sgd_decay_history, mini_batch_sgd_decay_training_time = train_model(mini_batch_sgd_decay_model, None, x_train, y_train, x_val, y_val, epochs=50, batch_size=64)

# Calculate metrics for Mini-batch SGD with decay
mini_batch_sgd_decay_predicted_labels = mini_batch_sgd_decay_model.predict(x_test)
mini_batch_sgd_decay_predicted_labels_flat = np.argmax(mini_batch_sgd_decay_predicted_labels, axis=1)
mini_batch_sgd_decay_f1, mini_batch_sgd_decay_precision, mini_batch_sgd_decay_recall = calculate_metrics(y_test_flat, mini_batch_sgd_decay_predicted_labels_flat)

plot_learning_curves(mini_batch_sgd_decay_history, 'Mini-batch SGD with Decay (Batch Size: 64)', color='Orange')
print("Training time for Mini-batch SGD with decay: {:.2f} seconds".format(mini_batch_sgd_decay_training_time))
print("Metrics for Mini-batch SGD with decay:")
print("F1 Score:", mini_batch_sgd_decay_f1)
print("Precision:", mini_batch_sgd_decay_precision)
print("Recall:", mini_batch_sgd_decay_recall)
```

```python
# Train the model with SGD with decay and momentum
sgd_decay_momentum_model = tf.keras.models.clone_model(model)
sgd_decay_momentum_model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=lr_schedule, momentum=0.9),
                                 loss='categorical_crossentropy',
                                 metrics=['accuracy'])
sgd_decay_momentum_history, sgd_decay_momentum_training_time = train_model(sgd_decay_momentum_model, None, x_train, y_train, x_val, y_val, epochs=50, batch_size=64)

# Calculate metrics for mini-batch SGD with decay and momentum
sgd_decay_momentum_predicted_labels = sgd_decay_momentum_model.predict(x_test)
sgd_decay_momentum_predicted_labels_flat = np.argmax(sgd_decay_momentum_predicted_labels, axis=1)
sgd_decay_momentum_f1, sgd_decay_momentum_precision, sgd_decay_momentum_recall = calculate_metrics(y_test_flat, sgd_decay_momentum_predicted_labels_flat)
plot_learning_curves(sgd_decay_momentum_history, 'SGD with Decay and Momentum', color='black')

print("Training time for SGD with decay and momentum: {:.2f} seconds".format(sgd_decay_momentum_training_time))
print("Metrics for SGD with decay and momentum:")
print("F1 Score:", sgd_decay_momentum_f1)
print("Precision:", sgd_decay_momentum_precision)
print("Recall:", sgd_decay_momentum_recall)
```

```python
# Train the model with Adam optimizer (lr=0.001)
adam_model = tf.keras.models.clone_model(model)
adam_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])
adam_history, adam_training_time = train_model(adam_model, None, x_train, y_train, x_val, y_val, epochs=50, batch_size=len(x_train))


# Plot learning curves for Adam
plot_learning_curves(adam_history, 'Adam (lr=0.001)', color='black')

# Print training times
print("Training time for Adam: {:.2f} seconds".format(adam_training_time))

# Calculate metrics for Adam
adam_predicted_labels = adam_model.predict(x_test)
adam_predicted_labels_flat = np.argmax(adam_predicted_labels, axis=1)
adam_f1, adam_precision, adam_recall = calculate_metrics(np.argmax(y_test, axis=1), adam_predicted_labels_flat)

print("Metrics for Adam:")
print("F1 Score:", adam_f1)
print("Precision:", adam_precision)
print("Recall:", adam_recall)
```

```python
# Train the model with RMSprop optimizer (lr=0.001)
rmsprop_model = tf.keras.models.clone_model(model)
rmsprop_model.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.001),
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])
rmsprop_history, rmsprop_training_time = train_model(rmsprop_model, None, x_train, y_train, x_val, y_val, epochs=50, batch_size=len(x_train))
plot_learning_curves(rmsprop_history, 'RMSprop (lr=0.001)', color='red')
print("Training time for RMSprop: {:.2f} seconds".format(rmsprop_training_time))

# Calculate metrics for RMSprop
rmsprop_predicted_labels = rmsprop_model.predict(x_test)
rmsprop_predicted_labels_flat = np.argmax(rmsprop_predicted_labels, axis=1)
rmsprop_f1, rmsprop_precision, rmsprop_recall = calculate_metrics(np.argmax(y_test, axis=1), rmsprop_predicted_labels_flat)

print("Metrics for RMSprop:")
print("F1 Score:", rmsprop_f1)
print("Precision:", rmsprop_precision)
print("Recall:", rmsprop_recall)
```

```python
# Export the best model trained with mini-batch SGD with decay and momentum
best_model_sgd_decay_momentum_filename = 'best_model_sgd_decay_momentum.h5'
sgd_decay_momentum_model.save(best_model_sgd_decay_momentum_filename)
import numpy as np
import tensorflow as tf

# Load the best model trained with mini-batch SGD with decay and momentum
best_model_sgd_decay_momentum = tf.keras.models.load_model('best_model_sgd_decay_momentum.h5')

# Make predictions on some samples from the test set
predictions = best_model_sgd_decay_momentum.predict(x_test[:10])  # Predictions on the first 10 samples

# Print the predicted labels
predicted_labels = np.argmax(predictions, axis=1)
print("Predicted labels:", predicted_labels)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define a function to display images along with their true labels and predicted labels
def display_images_with_labels(images, true_labels, predicted_labels):
    plt.figure(figsize=(12, 6))
    for i in range(10):
        plt.subplot(2, 5, i + 1)
        plt.imshow(images[i].reshape(28, 28), cmap='gray')
        plt.title(f'True: {true_labels[i]}, Predicted: {predicted_labels[i]}')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Display some images from the test set along with their true labels and predicted labels
display_images_with_labels(x_test[:10], np.argmax(y_test[:10], axis=1), predicted_labels[:10])
```
